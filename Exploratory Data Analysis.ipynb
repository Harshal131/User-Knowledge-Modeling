```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Dataset/Data.xls',sheet_name="Training_Data") 
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 6','Unnamed: 7','Attribute Information:'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.rename({' UNS': 'UNS'}, axis=1, inplace=True)
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
corr=df.corr()
sns.heatmap(corr,vmax=.3, center=0, cmap=cmap,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
sns.pairplot(df,diag_kind='kde',palette="husl")
```

```python
X=df[['STG','SCG','STR','LPR','PEG']]
Y=df['UNS']
```

```python
print(X.shape)
print(Y.shape)
```

```python
UNS=pd.value_counts(df['UNS'],sort=True)
UNS.plot(kind='bar')
```

```python
# Oversampling
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
```

```python
very_low=df[df['UNS']==0]
Low=df[df['UNS']==1]
Middle=df[df['UNS']==2]
High=df[df['UNS']==3]
```

```python
X_resampled, y_resampled = ros.fit_resample(X, Y)
```

```python
from collections import Counter
print(Counter(y_resampled))
```

```python
X_resampled.head()
```

```python
y_resampled.head()
```

```python
col=['STG','SCG','STR','LPR','PEG']
for i in col:
    plt.title('Distribution of {} column'.format(i))
    X_resampled[i].hist()
    plt.show()
```

```python
for i in col:
    plt.title(' Outlier detection of {} '.format(i))
    plt.boxplot(X_resampled[i])
    plt.show()
```

```python
sns.jointplot(x = 'STG', y = 'SCG', data = X_resampled, kind = 'hex', gridsize = 20);
sns.jointplot(x = 'SCG', y = 'STR', data = X_resampled, kind = 'hex', gridsize = 20);
sns.jointplot(x = 'STR', y = 'LPR', data = X_resampled, kind = 'hex', gridsize = 20);
sns.jointplot(x = 'LPR', y = 'PEG', data = X_resampled, kind = 'hex', gridsize = 20);
```

```python
X_resampled.to_csv(r'Dataset/X_train.csv', index = False)
y_resampled.to_csv(r'Dataset/Y_train.csv', index = False)
```
